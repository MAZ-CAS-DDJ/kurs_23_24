Vom Long- ins Wide-Format und umgekehrt

```python
import pandas as pd
```

### Daten

```python
df = pd.read_excel("dataprojects/widelong/kantone.xlsx")
```

```python
df
```

### Long to Wide

Dafür benutzen wir `df.pivot()`
- `index=`=> aus welcher Spalte wird der Index gebildet
- `columns=`=> aus welcher Spalte werden die Spalten gebildet
- `values=`=> aus welcher Spalte werden der jeweiligen Werte abgelesen

Referenz: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pivot.html

```python
df_wide = df.pivot(index="datum", columns="kanton", values="wert")
```

```python
df_wide
```

Verschönern:

```python
df_wide.index.name = None
```

```python
df_wide.columns.name = None
```

```python
df_wide
```

### Wide to Long

Hier ist die Syntax etwas länger. Herzstück ist `unstack()`.

Referenz: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.unstack.html

Das produziert erstmal eine Series:

```python
df_wide.unstack()
```

Daraus machen wir ein Dataframe:

```python
df_long = df_wide.unstack().to_frame().reset_index()
```

```python
df_long
```

Neue Spaltennamen definieren:

```python
df_long.columns = ["kanton", "datum", "wert"]
```

```python
df_long
```
