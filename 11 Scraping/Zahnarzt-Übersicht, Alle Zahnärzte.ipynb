# Zahnarztübersicht

Folgende Datenpunkte brauchen wir:
- Arztvergleich https://www.docapp.ch
- Name
- Typ/Titel
- Adresse
- Sterne
- Besuche
- Bewertungen

```python
#Imports
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Lösung mit Range-Iteration

```python
#Den HTML-Code mitnehmen
url = "https://www.docapp.ch/de/suche?q=Zahnarzt&l=Z%C3%BCrich&s=specialfield%20&page=1%20&sort=Relevance%20&size=25"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
zahnärzte = []
x = range(24)
for elem in x:
    Name = soup.find_all(class_="searchResult_text")[elem].find("a").text.strip()
    Titel = soup.find_all(class_="searchResult_text")[elem].find(class_="searchResult_title").text.strip()
    Aufrufe = soup.find_all(class_="searchResult_add")[elem].find("p").text.strip().replace("Profilbesuche", "")
    try:
        Bewertungen = soup.find_all(class_="searchResult_add")[elem].find("a").text.replace(" Bewertungen anzeigen >", "")
    except:
        Bewertungen = "N/A"
    Adresse = soup.find_all(class_="searchResult_adresse")[elem].text.strip()
    Sterne = soup.find(class_='docapp_searchResult_star').text
    mini_dict = {'Name': Name,
                 'Titel': Titel,
                 'Aufrufe': Aufrufe,
                 'Adresse': Adresse,
                 'Bewertungen': Bewertungen,
                 'Sterne': Sterne}
    zahnärzte.append(mini_dict)
```

```python
pd.DataFrame(zahnärzte)
```

## Alle Zahnärzte

```python
#Den HTML-Code mitnehmen
url = "https://www.docapp.ch/de/suche?q=Zahnarzt&l=Z%C3%BCrich&s=specialfield%20&page=1%20&sort=Relevance%20&size=100"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
zahnärzte_alle = []

for page in range(1,35):
    
    #Den HTML-Code mitnehmen
    url = "https://www.docapp.ch/de/suche?q=Zahnarzt&l=Z%C3%BCrich&s=%20&page="+str(page)+"%20&sort=Relevance%20&size=25"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    zahnärzte = []
    x = range(25)
    for elem in x:
        try:
            Name = soup.find_all(class_="searchResult_text")[elem].find("a").text.strip()
        except:
            Name = None
        try:
            Titel = soup.find_all(class_="searchResult_text")[elem].find(class_="searchResult_title").text.strip()
        except:
            Titel = None
        try:
            Aufrufe = soup.find_all(class_="searchResult_add")[elem].find("p").text.strip().replace("Profilbesuche", "")
        except:
            Aufrufe = None
        try:
            Bewertungen = soup.find_all(class_="searchResult_add")[elem].find("a").text.replace(" Bewertungen anzeigen >", "")
        except:
            Bewertungen = None
        try:
            Adresse = soup.find_all(class_="searchResult_adresse")[elem].text.strip()
        except:
            None
        try:
            Sterne = soup.find(class_='docapp_searchResult_star').text
        except:
            None
            
        mini_dict = {'Name': Name,
                 'Titel': Titel,
                 'Aufrufe': Aufrufe,
                 'Adresse': Adresse,
                 'Bewertungen': Bewertungen,
                 'Sterne': Sterne}
        
        zahnärzte_alle.append(mini_dict)
```

```python
pd.DataFrame(zahnärzte_alle).to_csv("Test.csv")
```
