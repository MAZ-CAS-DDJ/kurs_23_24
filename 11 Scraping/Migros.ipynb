```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.migros.ch/de/category/snacks-sussigkeiten/schokolade-sussigkeiten/pralinen'
```

```python
driver = webdriver.Firefox()
```

```python
driver.get(url)
```

```python
hersteller = driver.find_elements(By.CLASS_NAME, 'name')
for firma in hersteller:
    print(firma.text)
```

```python
artikels= driver.find_elements(By.CLASS_NAME, 'desc')
for artikel in artikels:
    print(artikel.text)
```

```python
preise = driver.find_elements(By.CLASS_NAME, 'actual')
for preis in preise:
    print(preis.text)
```

```python
gewichte = driver.find_elements(By.CLASS_NAME, 'weight-priceUnit')
for gewicht in gewichte:
    print(gewicht.text.replace("g", ""))
```

```python
final_list = []

for Hersteller, Artikel, Preis, Gewicht in zip (hersteller, artikels, preise, gewichte):
    
    minidict = {'Hersteller': Hersteller.text,
                'Produkt': Artikel.text,
                'Preis': float(Preis.text.replace('–','00')),
                'Gewicht': Gewicht.text.replace('g','')}
    
    final_list.append(minidict)
```

```python
pd.set_option('display.max_rows', None)
df= pd.DataFrame(final_list)
```

```python
import numpy as np
df = df.fillna(np.nan)
df = df.replace('', np.nan)
```

```python
df.sort_values('Preis', ascending=False)
```
