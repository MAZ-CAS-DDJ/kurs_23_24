# Zahnarztübersicht

Folgende Datenpunkte brauchen wir:
- Arztvergleich https://www.docapp.ch
- Name
- Typ/Titel
- Adresse
- Sterne
- Besuche
- Bewertungen

```python
#Imports
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Den HTML-Code mitnehmen
url = "https://www.docapp.ch/de/suche?q=Zahnarzt&l=Z%C3%BCrich&s=specialfield"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
# Typ/Titel
soup.find_all(class_="searchResult_text")[0].find(class_='searchResult_title').text.strip()
```

```python
# Name
soup.find_all(class_="searchResult_text")[0].find('a').text
```

```python
# Adresse
soup.find_all(class_="searchResult_adresse")[0].text.strip().split("\n")[0].strip()
```

```python
# Profilbesuche
int(soup.find_all(class_="searchResult_add")[0].find('p').text.strip().replace("Profilbesuche", ""))
```

```python
# Sterne
soup.find_all(class_="searchResult_add")[0].find(class_='docapp_searchResult_star').text
```

```python
# Bewertungen
soup.find_all(class_="docapp_searchResult_rating_box")[0].find("a").text.replace(" Bewertungen anzeigen >", "")
```

## Lösung mit unterschiedlichen Listen

```python
final_list = []

soup_txt = soup.find_all(class_="searchResult_text")
soup_adr = soup.find_all(class_="searchResult_adresse")
soup_num = soup.find_all(class_="searchResult_add")
soup_bew = soup.find_all(class_="docapp_searchResult_rating_box")


for txt, adr, num, bew in zip(soup_txt, soup_adr, soup_num, soup_bew):
    
    titel = txt.find(class_='searchResult_title').text.strip()
    name = txt.find('a').text
    adresse = adr.text.strip()
    besuche = num.find('p').text.strip().replace("Profilbesuche", "")
    Sterne = num.find(class_='docapp_searchResult_star').text
    bew = bew.find("a").text.replace(" Bewertungen anzeigen >", "")
    
    mini_dict = {'Titel': titel,
                'Name': name,
                'Adresse': adresse,
                'Besuche': besuche,
                'Sterne': Sterne,
                'Bewertungen': bew}
    
    final_list.append(mini_dict)
    
```

```python
pd.DataFrame(final_list)
```

# Lösung mit Range-Iteration

```python
zahnärzte = []
x = range(0,25)
for elem in x:
    
    Name = soup.find_all(class_="searchResult_text")[elem].find("a").text.strip()
    Titel = soup.find_all(class_="searchResult_text")[elem].find(class_="searchResult_title").text.strip()
    Aufrufe = soup.find_all(class_="searchResult_add")[elem].find("p").text.strip().replace("Profilbesuche", "")
    Bewertungen = soup.find_all(class_="searchResult_add")[elem].find("a").text.replace(" Bewertungen anzeigen >", "")
    Adresse = soup.find_all(class_="searchResult_adresse")[elem].text.strip()
    Sterne = soup.find(class_='docapp_searchResult_star').text
    
    mini_dict = {'Name': Name,
                 'Titel': Titel,
                 'Aufrufe': Aufrufe,
                 'Adresse': Adresse,
                 'Bewertungen': Bewertungen,
                 'Sterne': Sterne}
    
    zahnärzte.append(mini_dict)
```

```python
pd.DataFrame(zahnärzte)
```
