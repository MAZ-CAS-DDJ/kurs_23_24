# APIs bedienen (Lösung)

Wir machen nun die gleiche Abfrage wie vorhin, aber nutzen direkt die API. 

```python
import requests
import pandas as pd
```

```python
url = "https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2023-10-01&endtime=2023-10-04&orderby=magnitude&limit=10"

# Hier holen wir die Daten
response = requests.get(url)
data = response.json()

# Und nun sortieren wir nur das heraus, was uns interessiert
earthquakes = data['features']

final_list = []
for quake in earthquakes:
    magnitude = quake['properties']['mag']
    location = quake['properties']['place']
    latitude = quake['geometry']['coordinates'][1]
    longitude = quake['geometry']['coordinates'][0]
    depth = quake['geometry']['coordinates'][2]
    
    mini_dict = {"Magnitude": magnitude, 
                 "Location": location, 
                 "Latitude": latitude, 
                 "Longitude": longitude, 
                 "Depth": depth}
    
    final_list.append(mini_dict)
    
```

```python
pd.DataFrame(final_list)
```
