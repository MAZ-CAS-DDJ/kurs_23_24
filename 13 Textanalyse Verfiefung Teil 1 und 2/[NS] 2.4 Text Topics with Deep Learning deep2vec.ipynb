```python
import pandas as pd
df = pd.read_csv("data/tweets.csv")
```

```python
#list(df["text"])
```

```python
from top2vec import Top2Vec
documents = list(df["text"])
model = Top2Vec(documents)
```

```python
model.get_num_topics()
```

```python
topic_sizes, topic_nums = model.get_topic_sizes()
```

```python
pd.DataFrame([topic_sizes,topic_nums])
```

```python
topic_words, word_scores, topic_nums = model.get_topics(449)
```

```python
pd.DataFrame(topic_words)
```

```python
documents, document_scores, document_ids = model.search_documents_by_topic(topic_num=0, num_docs=5)
```
