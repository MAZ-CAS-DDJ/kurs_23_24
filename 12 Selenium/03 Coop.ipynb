```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
import pandas as pd
from bs4 import BeautifulSoup
```

# Untersuchung "Shrinkflation". Fokus Schoko-Pralines.
- Hersteller
- Produkte-Titel
- Preis
- Preis/100g

```python
url = "https://www.coop.ch/de/lebensmittel/suesses-snacks/schokolade-suessigkeiten/schokoladen-pralinen/c/m_0193"
```

```python
#Wir starten den BRowser auf
driver = webdriver.Firefox()
```

```python
#Und nun sagen wir dem Browser, welche Seite er besuchen sollte. 
driver.get(url)
```

```python
search = driver.find_element(By.CLASS_NAME, 'productTile__productMeta-value-item')
```

```python
search.text
```

```python
search = driver.find_elements(By.CLASS_NAME, 'productTile__productMeta-value-item')
```

```python
for elem in search:
    print(elem.text)
```

```python
company_names = driver.find_elements(By.CLASS_NAME, 'productTile__productMeta-value-item')
product_names = driver.find_elements(By.CLASS_NAME, 'productTile-details__name-value')
prices = driver.find_elements(By.CLASS_NAME, 'productTile__price-value')[0::2]
```

```python
final_list = []

for company, product, price in zip(company_names, product_names, prices):
    
    minidict = {'Hersteller': company.text,
                'Produkt': product.text,
                'Preis/100g': price.text}
    
    final_list.append(minidict)
```

```python
pd.DataFrame(final_list)
```

# Alles zusammen

```python
url = "https://www.coop.ch/de/lebensmittel/suesses-snacks/schokolade-suessigkeiten/schokoladen-pralinen/c/m_0193?page="

final_list = []

for elem in range (1, 5):
    
    elem = str(elem)
    url2 = url + elem
    
    print(url2)
    
    driver = webdriver.Firefox()
    driver.get(url2)
    
    company_names = driver.find_elements(By.CLASS_NAME, 'productTile__productMeta-value-item')
    product_names = driver.find_elements(By.CLASS_NAME, 'productTile-details__name-value')
    prices = driver.find_elements(By.CLASS_NAME, 'productTile__price-value')[0::2]
    
    for company, product, price in zip(company_names, product_names, prices):
    
        minidict = {'Hersteller': company.text,
                'Produkt': product.text,
                'Preis/100g': price.text}
    
        final_list.append(minidict)
    
    driver.close()
    
```

```python
pd.DataFrame(final_list)
```
