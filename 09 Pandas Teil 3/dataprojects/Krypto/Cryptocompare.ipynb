## Top 100 Currencies

```python
url = 'https://min-api.cryptocompare.com/data/top/mktcapfull'

parameters = {
    'limit': '100',
    'tsym': 'USD',
    'api_key': key
}
headers = {
    
}

session = Session()
session.headers.update(headers)

try:
    response = session.get(url, params=parameters)
    data = json.loads(response.text)
    #print(data)

except (ConnectionError, Timeout, TooManyRedirects) as e:
    data = ""
    #print(e)
```

```python
data
```

```python
len(data['Data'])
```

```python
currencies = []
```

```python
i = 0
```

```python
for d in data['Data']:
    print(i, d['CoinInfo']['Name'])
    currency = {}
    currency['id'] = d['CoinInfo']['Id']
    currency['name'] = d['CoinInfo']['Name']
    currency['fullname'] = d['CoinInfo']['FullName']
    currency['internal'] = d['CoinInfo']['Internal']
    try: 
        currency['marketcap'] = d['RAW']['USD']['MKTCAP']
    except:
        currency['marketcap'] = 0
    currencies.append(currency)
    i = i + 1
```

```python
df_currencies = pd.DataFrame(currencies)
```

```python
df_currencies
```

```python
df_currencies.to_csv("100_currencies_new.csv", index=False)
```

#### Execute with all currencies

Note: We only use 30!

```python
for currency in df['name'][0:30]:
    url = "https://min-api.cryptocompare.com/data/v2/histoday"
    
    parameters = {
        'fsym': currency,
        'tsym': 'USD',    
        'limit': '1826',
        'api_key': key
    }
    
    headers = {
    }

    session = Session()
    session.headers.update(headers)
    
    print(currency)

    try:
        response = session.get(url, params=parameters)
        data = json.loads(response.text)
        
        if data['Response'] != "Error":
            df_price = pd.DataFrame(data['Data']['Data'])
            df_price['date'] = pd.to_datetime(df_price['time'], unit="s")
            df_price = df_price[['date', 'open', 'high', 'low', 'close', 'volumeto']]
            df_price = df_price.rename(columns={'volumeto': 'volume'})
            df_price['currency'] = name_fullname[currency]
            df_price.to_csv("data/" + currency + ".csv", index=False)
        
            print(" - OK")
            
        else:
            print (" - Error")

    except (ConnectionError, Timeout, TooManyRedirects) as e:
        data = ""
        print(" - Error") 
        
    time.sleep(2)
```
